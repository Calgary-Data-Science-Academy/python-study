## 1. Sales Dataset

```python
import pandas as pd
import numpy as np
```

```python
sales = pd.read_csv('dataset/sales.csv',header=0,encoding ='latin1')
```

```python
sales.head()
```

make a copy of the original dataset and modify sales2

```python
sales2 = sales.copy()
```

```python
print(sales2.columns.values)
```

set the maximum display column numbers to 50

```python
pd.set_option('display.max_columns', 50)
```

```python
sales2.head()
```

Delete columns which are useless or have to many NAs

```python
to_drop = ['Unnamed: 0','gndr_cd', 'age_yr', 'income_amt', 'edu_val',
 'edu_1st_indv_val', 'edu_2nd_indv_val', 'adults_in_hh_num', 'married_ind',
 'child_present_ind', 'home_owner_ind', 'occpn_val', 'occpn_1st_val',
 'occpn_2nd_val', 'dist_to_ven']
sales2.drop(to_drop, inplace=True, axis=1)
```

```python
sales2.head()
```

Convert String to datetime format

```python
event_date_time=pd.to_datetime(sales2.event_date_time, errors='coerce')
```

```python
sales2 = sales2.assign(event_date=event_date_time.dt.date, event_time=event_date_time.dt.time)
```

Try to automatically find datetime columns and conver to datetime format

```python
mask = sales2.dropna().astype(str).apply(lambda x : x.str.match(r'\d{4}-\d{2}-\d{2} \d{2}\:\d{2}\:\d{2}').all())
sales2.loc[:,mask] = sales2.loc[:,mask].apply(pd.to_datetime)
mask2 = sales2.astype(str).apply(lambda x : x.str.match(r'\d{4}-\d{2}-\d{2}').all())
sales2.loc[:,mask2] = sales2.loc[:,mask2].apply(pd.to_datetime)
mask3 = sales2.astype(str).apply(lambda x : x.str.match(r'\d{2}\:\d{2}\:\d{2}').all())
sales2.loc[:,mask3] = sales2.loc[:,mask3].apply(pd.to_datetime)
```

The automatic searching did not work quite well due to certain columns have NULL values. You can also pick certain columns that you want to convert to datetime format as below:

```python
sales2[['event_date_time', 'event_dt', 'presale_dt','onsale_dt','sales_ord_create_dttm','sales_ord_tran_dt','print_dt']] =sales2[['event_date_time', 'event_dt', 'presale_dt','onsale_dt','sales_ord_create_dttm','sales_ord_tran_dt','print_dt']].applymap(pd.to_datetime)
```

```python
sales2.dtypes
```

```python
sales2.head()
```

Parse out a time stamp column into seperate columns with date and time

```python
sales2=sales2.assign(sales_ord_date=sales2['sales_ord_create_dttm'].dt.date, sales_ord_time=sales2['sales_ord_create_dttm'].dt.time)
```

```python
sales2.head()
```

Combine two string columns into one new column

```python
sales2['venue_city_state'] = sales2['venue_city'] +','+ sales2['venue_state']
```

```python
sales2['venue_city_state'].head()
```

## 2. MBTA Excel Dataset

use the pd.read_excel to read the mbta.xlsx file using default settings

```python
mbta = pd.read_excel('dataset/mbta.xlsx')
```

```python
mbta.head()
```

We do not need the first rows by observing the dataset. Set the column header as None

```python
mbta2 = pd.read_excel('dataset/mbta.xlsx',skiprows=1,header=None,
...               na_values=['NA', ''])
```

```python
mbta2.head()
```

```python
print(mbta2.columns.values)
```

Drop the first Column which is the index column in the original excel file

```python
mbta3=mbta2.drop([0],axis=1)
```

```python
mbta3.head()
```

Transpose the wide format data in mbta3 to the desired long format and use the first column as the header name

```python
mbta4=mbta3.set_index([1]).T
```

```python
mbta4.dtypes
```

Convert the column 'mode' to datetime format

```python
mbta4['mode']=pd.to_datetime(mbta4.mode, errors='coerce')
mbta4.describe()
```

Convert the rest of the columns to numeric format

```python
cols = mbta4.columns.drop('mode')
mbta4[cols] = mbta4[cols].apply(pd.to_numeric, errors='coerce')
mbta4.dtypes
```

```python
mbta4.describe()
```
